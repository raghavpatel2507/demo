```python
df=pd.read_excel("/content/chandan.xlsx")
```

```python
df.head()
```

```python
import category_encoders as ce
```

```python
encoder = ce.TargetEncoder(cols=['AccountName', 'ItemName', 'ColorName'])
```

```python
data_encoded = encoder.fit_transform(df[['AccountName', 'ItemName', 'ColorName']], df['TotalAmt'])

# Show the encoded data
print(data_encoded)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor
from sklearn.svm import SVR
```

```python
df.info()
```

```python
df['AccountName']=data_encoded['AccountName']
```

```python
df['ItemName']=data_encoded['ItemName']
df['ColorName']=data_encoded['ColorName']
```

```python
df.head()
```

```python
df.drop('DesignNo',axis=1,inplace=True)
```

```python
df.drop(['Size1','SalesRate'],axis=1,inplace=True)
```

```python
df.head(2)
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df.dropna(inplace=True)
```

```python
df.isnull().sum()
```

```python
df.duplicated().sum()
```

```python
df.drop_duplicates(inplace=True)
```

```python
df.duplicated().sum()
```

```python
df['VoucherDate']=pd.to_datetime(df['VoucherDate'])
```

```python
df['Year'] = df['VoucherDate'].dt.year
df['Month'] = df['VoucherDate'].dt.month
df['Day'] = df['VoucherDate'].dt.day
df.drop('VoucherDate',axis=1,inplace=True)
```

```python
X=df.drop('TotalAmt',axis=1)
y=df['TotalAmt']
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
```

```python
X_train.head()
```

```python
y_train
```

```python
lr=LinearRegression()
lr.fit(X_train, y_train)
```

```python
y_pred=lr.predict(X_test)
```

```python
y_pred
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
```

```python
mse = mean_squared_error(y_test, y_pred)
mae = mean_absolute_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

# Print results
print(f'Mean Squared Error (MSE): {mse}')
print(f'Mean Absolute Error (MAE): {mae}')
print(f'R-squared (R2): {r2}')


print(f'Intercept: {lr.intercept_}')
print(f'Coefficient: {lr.coef_}')
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
actual_sales = y_test
predicted_sales = y_pred
data = pd.DataFrame({
    'Actual Sales': actual_sales,
    'Predicted Sales': predicted_sales
})
plt.figure(figsize=(8, 6))
sns.scatterplot(x='Actual Sales', y='Predicted Sales',data=data)
plt.plot([min(actual_sales), max(actual_sales)], [min(actual_sales), max(actual_sales)], color='red', linestyle='--', label='Perfect Prediction (y = x)')
plt.grid(True, linestyle='--', alpha=0.7)
plt.title('Actual vs Predicted Sales', fontsize=16)
plt.xlabel('Actual Sales', fontsize=14)
plt.ylabel('Predicted Sales', fontsize=14)
plt.legend()


plt.show()
```
